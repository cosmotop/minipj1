## 3. 데이터 전처리

전처리 과정에서 다음과 같은 처리를 할 수 있습니다.

- 결측치 처리
- 값 변경
- 열 추가
- 불필요한 열 제거
- 열 이름 변경
- 데이터 통합(연결 또는 조인)
- 가변수화 등

**1) 데이터 통합**

- concat을 사용해 연결하면 인덱스 값을 기준으로 연결되어 데이터가 어긋납니다.
- **year** 열을 기준으로 **outer** 방식으로 조인(병합)합니다.
- 외국인 정보가 1981 ~ 1984년이 누락되어 결측치가 발생할 것입니다.
- 세 개의 데이터프레임을 병합(조인)하여 pop 데이터프레임을 선언합니다.
- 이후의 모든 작업은 pop 데이터프레임을 대상으로 진행합니다.

```python
# 데이터프레임 조인
pop=pd.merge(pop01,pop02,how='outer')
pop=pd.merge(pop,pop03,how='outer')

# 확인
pop
```

**2) 결측치 확인**

- 결측치가 있는지 확인합니다.

```python
# 결측치 확인
pop.isnull().sum()
```

**3) 결측치 저리**

- 연도별 인구 현황이므로 임의의 값을 채우는 것이 바람직하지 않아 보입니다.
- 이후 값, 즉 1985년 값으로 채우는 것도 정확한 분석을 방해할 것 같습니다.
- 이에 결측치가 있는 1981년~1984년 행을 제거할 것입니다.

```python
# 결측치 제거
pop.dropna(axis=0,inplace=True)
           
# 확인
pop
```

**3) 열 추가**

- 이후 분석의 편의를 위해 다음과 같은 의미를 갖는 열을 추가하고자 합니다.
    - k_total = 전체 한국인 인구수
    - f_total = 전체 외국인 인구수
    - male = 전체 남자 인구수
    - female = 전체 여자 인구수
    - total = 전체 인구수
- 추가할 열에 대한 공식은 다음과 같습니다.
    - k_total = k_male + k_female
    - f_total = f_male + f_female
    - male = k_male + f_male
    - female = k_female + f_female
    - total = k_total + f_total
- 정리한 공식에 따라 데이터프레임에 열을 추가합니다.

```python
# 열 추가
pop['k_total']=pop['k_male']+pop['k_female']
pop['f_total']=pop['f_male']+pop['f_female']
pop['male']=pop['k_male']+pop['f_male']
pop['female']=pop['k_female']+pop['f_female']
pop['total']=pop['k_total']+pop['f_total']

#  확인
pop
```

**4) (옵션) 열 순서 변경**

- 데이터 이해를 돕기 위해 다음과 같은 순서로 데이터프레임 열 순서를 변경합니다.
- year, household, total, male, female, k_total, k_male, k_female, f_total, f_male, f_female, older_65


```python
# 열 순서 변경
col = ['year','household','total','male','female', 'k_total', 'k_male','k_female', 'f_total', 'f_male', 'f_female', 'older_65']
pop = pop[col]

# 확인
pop
```

**5) (옵션) 인덱스 초기화**

- 인덱스가 0부터 시작하는 일련 변호를 갖지 않는다면 인덱스를 초기화합니다.

```python
# 인덱스 초기화
pop.reset_index(drop=True,inplace=True)

# 확인
pop
```

## 5.정리

- 분석을 통해 확인한 내용을 정리합니다.

**1) 의견**

- 전체 인구수 감소 추세, 특히 남성 인구수 빠르게 감소
- 가구당 인구수 감소 -> 1인 가구 증가
- 전체 인구에서 외국인이 차지하는 비중 늘어남
- 총 인구에서 노령 인구가 차지하는 비율 늘어남
- 여성 비율 증가

**2) 기타**

- 특별하게 분석 목적이 주어지지 않고 데이터만 보고 특정 아이디어를 만들어서 분   석을 진행해야 해 어려웠다.
- 서울의 외국인 인구수가 늘어나는 추세로 이에 대한 적절한 규제와 장려 정책이 요구된다.
- 1인 가구가 늘어나는 추세로 1인 가구 지원 정책을 늘릴 필요성이 있다.
- 노령화 인구가 늘어나고 있기 때문에 노령 인구 복지 정책을 확대해야 한다.
- 서울 인구가 한국 인구의 대부분이기 때문에 국내 인구 또한 인구수 감소와 1인가구 증가 추세를 가질 것으로 예상된다. 이는 결혼 비율과 출산률이 극도로 낮아지고  있기 때문으로 생각되어 진다. 노령 인구가 늘어나 필요한 복지 비용이 늘어나기 때  문에 인구 감소는 결혼, 출산 지원과 외국인 인구수 증대를 통해 일정 인구수를 유지하는 방향의 정책이 필요하다.
- 추가적인 데이터가 주어졌다면 정책의 방향성을 정할 수 있었을 것이다. 예를 들어 아이 1명을 키우는데 드는 평균 비용, 그 비용에서 교육이 차지하는 비율 등의 데이터가 존재했다면 새로운 교육지원 정책을 통해 출산율 증대를 노릴 수 있었을 것이다.
